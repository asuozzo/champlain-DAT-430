```python
import pandas as pd
```

### Process World Bank Files

```python
lifeexpectancy = pd.read_csv(
    "data/rawdata/life-expectancy-at-birth.csv",skiprows=2,header=1
)
lifeexpectancy_male = pd.read_csv(
    "data/rawdata/life-expectancy-at-birth-male.csv",skiprows=2, header=1
)
lifeexpectancy_female = pd.read_csv(
    "data/rawdata/life-expectancy-at-birth-female.csv",skiprows=2, header=1
)

lifeexpectancy = lifeexpectancy.append(
    [lifeexpectancy_male,lifeexpectancy_female],
    sort=False
)

lifeexpectancy.drop(["Unnamed: 63","2018","Indicator Code"],inplace=True,axis=1)

lifeexpectancy = pd.melt(
    lifeexpectancy,
    id_vars=[
        "Country Name","Country Code","Indicator Name",
    ],
    value_name="Life Expectancy",
    var_name="Year"
)
lifeexpectancy['Indicator Name'] = lifeexpectancy['Indicator Name'].str.replace(
    " \(years\)",""
).str.replace(
    "at birth,","-"
)

lifeexpectancy = lifeexpectancy.pivot_table(
    index=['Country Name','Country Code','Year'],
    columns=["Indicator Name"],
    values="Life Expectancy"
).reset_index()

lifeexpectancy.to_csv("data/wb-life-expectancy-at-birth.csv", index=False)
lifeexpectancy.head()
```

```python
netincome = pd.read_csv(
    "data/rawdata/net-income-per-capita.csv",skiprows=2,header=1
)
netincome.drop([
    "Indicator Code","Unnamed: 63","2018", 'Indicator Name'
], inplace=True,axis=1)
netincome = pd.melt(
    netincome,
    id_vars=['Country Name','Country Code'],
    value_name="Net Income",
    var_name="Year"
)

netincome.to_csv("data/wb-net-income.csv", index=False)
netincome.head()
```

```python
population = pd.read_csv(
    "data/rawdata/population.csv",skiprows=2,header=1
)
population.drop([
    "Indicator Code","Unnamed: 63","2018", 'Indicator Name'
], inplace=True,axis=1)
population = pd.melt(
    population,
    id_vars=['Country Name','Country Code'],
    value_name="Population",
    var_name="Year"
)

population.to_csv("data/wb-population.csv", index=False)
population.head()
```
